```python
import re

def classify_element(element):
    # 正则表达式模式
    pattern_id = r'^\d+$'  # 全数字（病人ID）
    pattern_code = r'^[A-Z][A-Z\d]*$'  # 大写字母开头，后跟数字or大写字母(icd10)
    pattern_protein = r'^[a-z\d][a-z\d_]*$'
    
    # 分类判断
    if re.match(pattern_id, element):
        return 'eid'
    elif re.match(pattern_code, element):
        return 'icd10'
    elif re.match(pattern_protein, element):
        return 'protein'
    else:
        return 'metabolite'
```

```python
keys_classification = [classify_element(key) for key in loaded_keys]
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances
from typing import List, Optional, Dict, Tuple
from scipy.stats import spearmanr

def get_knn_specific_types( # 不包含自己
    embeddings: np.ndarray,
    names: List[str],
    metric: str,
    k: int,
    source_type: str,
    target_type: str
) -> Dict[str, List[Tuple[str, float]]]:
    if metric=='spearman':
        corr_matrix, _ = spearmanr(embeddings, axis=1) #axis=1 means rows
        dists = 1.0 - np.abs(corr_matrix)
    else:
        dists = pairwise_distances(embeddings, metric=metric)
    n = len(names)
    result: Dict[str, List[Tuple[str, float]]] = {}
    for i, name in enumerate(names):
        name_class = classify_element(name)
        if name_class==source_type:
            dist_row = dists[i]
            sorted_idx = np.argsort(dist_row)
            neighbors: List[Tuple[str, float]] = []
            for j in sorted_idx:
                names_j = names[j]
                if classify_element(names_j) == target_type:
                    dist_ij = float(dist_row[j])
                    neighbors.append((names_j, dist_ij))
                if len(neighbors) >= k:
                    break
            result[name] = neighbors

    return result
```

```python
def adjust_node_order(df):
    """调整DataFrame中node1和node2的顺序，确保eid在node1，protein在node2"""
    # 复制DataFrame以避免修改原始数据
    adjusted_df = df.copy()
    
    # 遍历每一行，检查node1和node2的类型
    for index, row in adjusted_df.iterrows():
        node1_type = classify_element(row['node1'])
        node2_type = classify_element(row['node2'])
        
        # 如果node1是protein/metabolite且node2是eid，则交换它们
        if node1_type in ['protein', 'metabolite'] and node2_type == 'eid':
            adjusted_df.at[index, 'node1'], adjusted_df.at[index, 'node2'] = row['node2'], row['node1']
    
    return adjusted_df
```

```python
# 调整节点顺序
reorder_input_relationship = adjust_node_order(input_relationship)   
```

```python
reorder_input_relationship = reorder_input_relationship.drop(columns=['relationship'])
```

```python
def filter_specific_rows(df, node1_type, node2_type):
    # 复制DataFrame以避免修改原始数据
    filtered_df = df.copy()
    
    # 应用分类函数并筛选符合条件的行
    filtered_df['node1_type'] = filtered_df['node1'].apply(classify_element)
    filtered_df['node2_type'] = filtered_df['node2'].apply(classify_element)
    
    # 筛选node1为eid且node2为protein的行
    result_df = filtered_df[(filtered_df['node1_type'] == node1_type) & 
                            (filtered_df['node2_type'] == node2_type)]
    
    # 删除临时添加的类型列
    result_df = result_df.drop(columns=['node1_type', 'node2_type'])
    
    return result_df
```

```python
reorder_input_relationship_filter_eid_protein = filter_specific_rows(df=reorder_input_relationship,
                                                                     node1_type='eid',
                                                                     node2_type='protein')
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
def compare_dataframes(df1, df2, col1='node1', col2='node2'):
    """
    比较两个DataFrame的行，计算交集和差异
    
    参数:
    df1, df2: 需要比较的两个DataFrame
    col1, col2: 用于比较的列名，默认为'node1'和'node2'
    
    返回:
    包含比较结果的字典
    """
    # 将DataFrame的指定列组合成元组集合
    set1 = set(zip(df1[col1], df1[col2]))
    set2 = set(zip(df2[col1], df2[col2]))
    
    # 计算交集和差异
    intersection = set1 & set2
    only_in_df1 = set1 - set2
    only_in_df2 = set2 - set1
    
    # 创建结果DataFrame
    df_intersection = pd.DataFrame(list(intersection), columns=[col1, col2])
    df_only_in_df1 = pd.DataFrame(list(only_in_df1), columns=[col1, col2])
    df_only_in_df2 = pd.DataFrame(list(only_in_df2), columns=[col1, col2])
    
    return {
        'intersection': df_intersection,
        'only_in_df1': df_only_in_df1,
        'only_in_df2': df_only_in_df2,
        'counts': {
            'df1_total': len(df1),
            'df2_total': len(df2),
            'intersection': len(intersection),
            'only_df1': len(only_in_df1),
            'only_df2': len(only_in_df2)
        }
    }
```

```python
result_compare_dataframes = compare_dataframes(df1=eid_knn_protein_dataframe, 
                                               df2=reorder_input_relationship_filter_eid_protein,
                                               col1='node1', 
                                               col2='node2')
```

```python
def plot_venn_diagram(counts, labels=['eid_50NN_protein_output', 'eid_protein_input']):
    """
    绘制韦恩图展示两个DataFrame的比较结果
    
    参数:
    counts: 包含计数信息的字典
    labels: 韦恩图中两个集合的标签
    """
    # 创建画布
    plt.figure(figsize=(10, 8))
    
    # 绘制韦恩图
    venn2(
        subsets=(counts['only_df1'], counts['only_df2'], counts['intersection']),
        set_labels=labels,
        set_colors=('skyblue', 'lightgreen'),
        alpha=0.7
    )
    
    # 添加标题
    plt.title('', fontsize=14)
    
    # 显示图形
    plt.tight_layout()
    plt.show()
```

```python
# 打印统计结果
print("统计结果:")
print(f"eid_knn_protein_dataframe总行数: {result_compare_dataframes['counts']['df1_total']}")
print(f"reorder_input_relationship_filter_eid_protein总行数: {result_compare_dataframes['counts']['df2_total']}")
print(f"相同的行数: {result_compare_dataframes['counts']['intersection']}")
print(f"仅存在于eid_knn_protein_dataframe的行数: {result_compare_dataframes['counts']['only_df1']}")
print(f"仅存在于reorder_input_relationship_filter_eid_protein的行数: {result_compare_dataframes['counts']['only_df2']}")
```

```python
# 绘制韦恩图
plot_venn_diagram(result_compare_dataframes['counts'])
```
